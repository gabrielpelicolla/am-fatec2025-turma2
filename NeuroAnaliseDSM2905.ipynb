#Precisão do Neuronio de Arvore

```python
devedores = np.sum(y_credit_teste == 1)
```

```python
pa_a = pa
pb_a= devedores / len(y_credit_teste)
pa_a, pb_a
```

```python
pab_a=(pa_a * pb_a)/((pa_a* pb_a) + ((1-pa_a)*(1-pb_a)))
print(f'P(AB) = {pab_a *100:.2f}%')
```

#Precisão do Neuronio de Random Forest

```python
pa_r=rpa
pb_r=devedores/len(y_credit_teste)

pa_r, pb_r
```

```python
pab_r=(pa_r * pb_r)/((pa_r* pb_r) + ((1-pa_r)*(1-pb_r)))
print(f'P(AB) = {pab_r *100:.2f}%')
```
